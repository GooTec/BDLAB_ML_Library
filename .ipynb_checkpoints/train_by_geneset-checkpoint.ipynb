```python
import pandas as pd
import tensorflow as tf
import lib.dataProcess as dp
import lib.deepLearning as dl
import tensorboard as tb
```

```python
raw_data = pd.read_csv('./GEO_input_ensemble_CV_500.csv')
gene_sets = dp.gmt_to_dictionary('./c6.all.v6.1.symbols.gmt')
```

```python
raw_genes = raw_data.columns.values.tolist()
raw_genes = raw_genes[1:-3]
raw_dict = {}
for i in range(len(raw_genes)):
    raw_dict[raw_genes[i]] = i
raw_dict
```

```python
gene_idx = {}
for key, values in gene_sets.items() : 
    gene_idx[key] = []
    for value in values :
        if value in raw_dict.keys():
            gene_idx[key].append(raw_dict[value])
print(gene_idx)
```

```python
fivefold = dp.n_fold(raw_data, 'index', 5)
```

```python
xdata_five, ydata_five = dp.divide_xy_train(fivefold, 'result', True, 1, -3)
train_x, test_x = dp.train_test(xdata_five, 0)
train_y, test_y = dp.train_test(ydata_five, 0)
train_y = dp.one_hot_encoder(train_y)
test_y = dp.one_hot_encoder(test_y)
val_x, test_x = dp.test_validation(test_x)
val_y, test_y = dp.test_validation(test_y)
```

```python
sub_batches = []
batch_size = 100
train_loop = len(train_x)//100
for i in range(train_loop+1):
    sub_batch = []
    for geneset, idx in gene_idx.items() :
        if idx : 
            sub_batch.append(train_x[i*100:(i+1)*100, idx])
    sub_batch.append(train_y[i*100:(i+1)*100])
    sub_batch.append(0.7)
    sub_batch.append(True)
    sub_batches.append(sub_batch)
print(len(sub_batches) , len(sub_batches[0]))
```

```python
len(train_x)
```

```python
learning_rate  = 0.01
```

```python
subtrains = []
subtests = []
subvals = []
for geneset, idx in gene_idx.items() :
    if idx : 
        subtrain_x = train_x[:, idx]
        subtrains.append(subtrain_x)
        subtest_x = test_x[:, idx]
        subtests.append(subtest_x)
        subval_x = val_x[:, idx]
        subvals.append(subval_x)
tf.reset_default_graph()

placeholders = []
count = 0 
for geneset, idx in gene_idx.items():
    if idx : 
        placeholders.append(tf.placeholder(tf.float32, [None, len(subtrains[count][0])]))
        count += 1 
genesetlayer = []
count = 0

placeholders.append(tf.placeholder(tf.float32, [None, len(train_y[0])])) # Y tensor
placeholders.append(tf.placeholder(tf.float32)) # keep_prob
placeholders.append(tf.placeholder(tf.bool, name='phase')) # phase

for genesetname, idx in gene_idx.items():
    if idx : 
        genesetlayer.append(dl.dense(placeholders[count], 1, genesetname, tf.nn.relu))
        count += 1

concatenated_genesetlayer = tf.concat(genesetlayer, 1)     
h1 = dl.dense_dropout(concatenated_genesetlayer, placeholders[-1], 100, 'layer1', placeholders[-2] )
h2 = dl.dense_dropout(h1, placeholders[-1] , 50, 'layer2', placeholders[-2])
logits = dl.dense(h2, len(train_y[0]), 'logits', None)
hypothesis = tf.nn.softmax(logits)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits= logits, labels= placeholders[-3]))
train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

predicted = tf.argmax(hypothesis, 1)
correct_prediction = tf.equal( predicted, tf.argmax( placeholders[-3], 1))
accuracy = tf.reduce_mean(tf.cast( correct_prediction, dtype=tf.float32))
subtrains.append(train_y)
subtrains.append(1.0)
subtrains.append(False)
subvals.append(val_y)
subvals.append(1.0)
subvals.append(False)
subtests.append(test_y)
subtests.append(1.0)
subtests.append(False)
prediction_summary = tf.summary.tensor_summary('prediction', predicted)
concatenated_summary = tf.summary.tensor_summary('concatenated', concatenated_genesetlayer)
merged_summary = tf.summary.merge_all()


print(len(subtrains), len(subtrains[0]))
```

```python
best_train_acc = 0
best_val_acc = 0
best_test_acc = 0 
best_cost = float("inf")
save_path_dir ='./'
count = 0
model_num = 0
saver = tf.train.Saver()

with tf.Session() as sess:
    writer = tf.summary.FileWriter('test')
    writer.add_graph(sess.graph)

    sess.run(tf.global_variables_initializer())
#     sess.run(train, feed_dict={ i: d for i, d in zip(placeholders, subtrains)})
    summary = sess.run(merged_summary, feed_dict= { i: d for i, d in zip(placeholders, subtrains)})
#     summary,_=sess.run([merged_summary,train], feed_dict={i: d for i, d in zip(placeholders, subtrains)})
    writer.add_summary(summary)
    stop_switch = True
    step = 0

    while stop_switch:
        for i in range(train_loop):
            sess.run(train, feed_dict= { i: d for i, d in zip(placeholders, sub_batches[i])})


        train_h, train_c, train_p, train_a = sess.run( [hypothesis, cost, predicted, accuracy], feed_dict= { i: d for i, d in zip(placeholders, subtrains)})
        val_h, val_c, val_p, val_a = sess.run([hypothesis, cost, predicted, accuracy], feed_dict= { i: d for i, d in zip(placeholders, subvals)})
        if step % 1 == 0 :
            print("train acc : ", train_a, "validation acc : ", val_a, "train_cost", train_c)
        step += 1

        if best_cost > val_c :
            best_train_acc = train_a
            best_val_acc = val_a
            best_cost = val_c
            count = 0
            save_path = saver.save(sess, save_path_dir + 'model'+str(model_num)+'.ckpt')

        elif count > 10 :
            stop_switch = False
            print("Learning Finished!! \n")
        else:
            count += 1

    print("Training Accuracy : ", best_train_acc,  "Validation Accuracy : ", best_val_acc)

    saver.restore(sess, save_path)

    test_h, test_p, test_a = sess.run([hypothesis, predicted, accuracy],
                                      feed_dict= { i: d for i, d in zip(placeholders, subtests)})
    print("\nTest Accuracy: ", test_a)
    best_test_acc = test_a


    model_num += 1
```
